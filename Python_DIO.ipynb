```python
df = pd.read_excel("AdventureWorks Sales.xlsx", sheet_name="Sales_data")
```

```python
df.head()
```

```python
df.shape
df.dtypes
```

```python
# receita total
round(df["Sales Amount"].sum(),2)
```

```python
# custo total
df["Custo"] = df["Product Standard Cost"] * df["Order Quantity"]
```

```python
df.head()
```

```python
# tempo de envio
df["OrderDateKey"] = pd.to_datetime(df["OrderDateKey"], format="%Y%m%d") 
df["DueDateKey"] = pd.to_datetime(df["DueDateKey"], format="%Y%m%d") 
df["ShipDateKey"] = pd.to_datetime(df["DueDateKey"], format="%Y%m%d") 
df.head()
```

```python
df["tempo"] = (df["ShipDateKey"] - df["OrderDateKey"]).dt.days
df.head()
```

```python
df.dtypes
```

```python
df.groupby("ResellerKey")["tempo"].mean()
```

```python
df.isnull().sum()
```

```python
df.groupby([df["OrderDateKey"].dt.year,"ResellerKey"])["Sales Amount"].mean()
```

```python
pd.options.display.float_format = '{:20,.2f}'.format
df.groupby([df["OrderDateKey"].dt.year,"ResellerKey"])["Sales Amount"].sum()
```

```python
df.groupby("ProductKey")["Order Quantity"].sum().sort_values(ascending=False).plot.bar(title="Product Total");
```

```python
df.groupby(df["OrderDateKey"].dt.year)["Sales Amount"].sum().plot.bar("Sales Total")
plt.xlabel("Year")
plt.ylabel("US$")
plt.xticks(rotation='horizontal');
```

```python
from matplotlib.pyplot import xlabel
df_2019 = df[df["OrderDateKey"].dt.year == 2019]
df_2019.groupby(df_2019["OrderDateKey"].dt.month)["Sales Amount"].sum().plot(title="Sales Month")
```

```python
df_2019.groupby("ResellerKey")["Sales Amount"].sum().plot(title="Sales Month ReSeller")
```
